```python
# 선 색상 수정하기

plt.style.use("seaborn")

def mycal(num):
  num2 = num ** 2
  return num2

x = [1, 3, 5, 7, 9]
y = [2, 4, 3, 6, 10]
y2 = []

for x_each in x:
  y2.append(mycal(x_each))

plt.plot(x, y, color = "red")
plt.plot(x, y2, color = "darkcyan")

plt.show()
```

```python
# 연습문제 1

path = "/content/gdrive/MyDrive/pytest_basic/pytest_basic/"
data = pd.read_csv(path + 'boston_housing.csv')
# print(data.head())

# crim만 정렬
c = data.sort_values(by='crim', ascending=False)
list = c["crim"].iloc[9]
c.iloc[0:9, 0] = list
# print(c)

# age 80 이상인 값에 대해 crim 평균 구하기
a = c[c.age >= 80]
b = a['crim'].mean()  # mean() = 평균
print(b)
```

```python
# 연습문제 2

path = "/content/gdrive/MyDrive/pytest_basic/pytest_basic/"
data = pd.read_csv(path + 'california_housing.csv')

# print(data)

total = data['total_bedrooms']

# 데이터 추출
percent = len(data) * 0.8
# print(percent)

list = data.loc[0:16512]
# print(list)

# 중앙값 = median()
middle = data['total_bedrooms'].median()
print(middle)

# 결측치 치환
change = data.fillna({'total_bedrooms' : middle})
# change


# 표준편차 값의 차이의 절댓값
import numpy as np

# 표준편차
sq1 = np.std(total)
# total_bedrooms 값 바꾼 것을 표준편차 만들기
sq2 = np.std(change['total_bedrooms'])
# 표준편차 값의 차이 절댓값
print("표준편차 전후의 차이: ", sq1-sq2)
```

```python
# 연습문제 3

path = "/content/gdrive/MyDrive/pytest_basic/pytest_basic/"
data = pd.read_csv(path + 'insurance.csv')

d = data['charges']

# 평균
m = data['charges'].mean()
# 표준편차
s = np.std(d)

print(m, s)

# 이상 값 찾기
high = data[d >= m + s*1.5]

result = high['charges'].sum()
result
```
